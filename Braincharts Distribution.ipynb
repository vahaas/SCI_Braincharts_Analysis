```python
import pandas as pd
import matplotlib.pyplot as plt

# Excel-Datei laden
df = pd.read_excel('/Volumes/team/INANL/56_Valentina_Haas/brain_age_centilbrain/MasterThesis/Braincharts data/all_age_eLife.xlsx')  # oder .csv falls es eine CSV ist


# Alterswerte auf ganze Jahre runden
df['age_rounded'] = df['age'].round()

# Häufigkeit der gerundeten Alterswerte zählen
age_counts = df['age_rounded'].value_counts().sort_index()


# Plotten
plt.figure(figsize=(10,6))
plt.bar(age_counts.index, age_counts.values)
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Count per age')
plt.show()
```

```python
# Altersbereich 34 bis 75 auswählen
age_counts_filtered = age_counts.loc[34:75]

# Plotten
plt.figure(figsize=(10,6))
plt.bar(age_counts_filtered.index, age_counts_filtered.values)
plt.xlabel('Age (rounded)')
plt.ylabel('Count')
plt.title('Count per age (34–75)')
plt.show()
```
