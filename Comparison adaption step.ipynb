```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Dateien laden
path_orig = '/Volumes/team/INANL/56_Valentina_Haas/brain_age_centilbrain/07_BrainCharts/Adaption_Results_New/Rescaled/Braincharts_SCI_Template_06.csv'
path_adj = '/Volumes/team/INANL/56_Valentina_Haas/brain_age_centilbrain/07_BrainCharts/Adaption_Results_New/Rescaled/Adjusted_Values_Final.csv'
# CSVs laden
original = pd.read_csv(path_orig)
adjusted = pd.read_csv(path_adj)
# Funktion zum Bereinigen: Strings -> Float
def clean_numeric(df):
    # Alle Spalten außer 'subject' bereinigen
    for col in df.columns:
        if col != 'subject':
            # Entfernt Klammern und Leerzeichen, ersetzt Komma durch Punkt (falls nötig)
            df[col] = (
                df[col]
                .astype(str)
                .str.replace(r'[\[\]\s]', '', regex=True)
                .str.replace(',', '.', regex=False)
            )
            # In Float umwandeln, Fehler als NaN
            df[col] = pd.to_numeric(df[col], errors='coerce')
    return df
# Daten bereinigen
original = clean_numeric(original)
adjusted = clean_numeric(adjusted)

# Gemeinsame Features (außer subject)
common_features = list(set(original.columns).intersection(adjusted.columns))
common_features.remove('subject')

# Rename Columns
original_renamed = original.rename(columns={f: f"{f}_orig" for f in common_features})
adjusted_renamed = adjusted.rename(columns={f: f"{f}_adj" for f in common_features})

# Merge über subject
merged = pd.merge(original_renamed, adjusted_renamed, on='subject')

# Differenz berechnen
for feature in common_features:
    merged[f'{feature}_diff'] = merged[f'{feature}_adj'] - merged[f'{feature}_orig']

# Zusammenfassung: Mittelwert & Standardabweichung der Änderungen
summary = pd.DataFrame({
    'mean_diff': merged[[f'{f}_diff' for f in common_features]].mean(),
    'std_diff': merged[[f'{f}_diff' for f in common_features]].std()
})

summary.reset_index(inplace=True)
summary.rename(columns={'index': 'Feature'}, inplace=True)

# Ausgabe
print(summary)
# Speichern als Excel-Datei
output_path = '/Volumes/team/INANL/56_Valentina_Haas/brain_age_centilbrain/MasterThesis/Final_Version/Braincharts_Diff_Summary.xlsx'
summary.to_excel(output_path, index=False)

print(f'Excel-Datei gespeichert: {output_path}')
```
